```python
import matplotlib.pyplot as plt

resultados = {
    'Execução 1': {
        'Mochila Cheia': {'Resultado': 373, 'Tempo': 1.4e-06},
        'Substituição de Objeto': {'Resultado': 373, 'Tempo': 4.39e-05},
        'Hill Climbing': {'Resultado': 373, 'Tempo': 4.89e-05}
    },
    'Execução 2': {
        'Mochila Cheia': {'Resultado': 154, 'Tempo': 1.8e-06},
        'Substituição de Objeto': {'Resultado': 154, 'Tempo': 0.0001199},
        'Hill Climbing': {'Resultado': 154, 'Tempo': 9.01e-05}
    },
    'Execução 3': {
        'Mochila Cheia': {'Resultado': 496, 'Tempo': 2.4e-06},
        'Substituição de Objeto': {'Resultado': 496, 'Tempo': 1.55e-05},
        'Hill Climbing': {'Resultado': 496, 'Tempo': 1.74e-05}
    }
}

execucoes = ['Execução 1', 'Execução 2', 'Execução 3']
mochila_cheia_result = [resultados[exec]['Mochila Cheia']['Resultado'] for exec in execucoes]
substituicao_objeto_result = [resultados[exec]['Substituição de Objeto']['Resultado'] for exec in execucoes]
hill_climbing_result = [resultados[exec]['Hill Climbing']['Resultado'] for exec in execucoes]

mochila_cheia_tempo = [resultados[exec]['Mochila Cheia']['Tempo'] for exec in execucoes]
substituicao_objeto_tempo = [resultados[exec]['Substituição de Objeto']['Tempo'] for exec in execucoes]
hill_climbing_tempo = [resultados[exec]['Hill Climbing']['Tempo'] for exec in execucoes]

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(execucoes, mochila_cheia_result, label='Mochila Cheia', marker='o')
plt.plot(execucoes, substituicao_objeto_result, label='Substituição de Objeto', marker='o')
plt.plot(execucoes, hill_climbing_result, label='Hill Climbing', marker='o')
plt.title('Resultados por Execução')
plt.xlabel('Execução')
plt.ylabel('Resultado')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(execucoes, mochila_cheia_tempo, label='Mochila Cheia', marker='o')
plt.plot(execucoes, substituicao_objeto_tempo, label='Substituição de Objeto', marker='o')
plt.plot(execucoes, hill_climbing_tempo, label='Hill Climbing', marker='o')
plt.title('Tempos por Execução')
plt.xlabel('Execução')
plt.ylabel('Tempo (segundos)')
plt.legend()

plt.tight_layout()
plt.show()
```

Os algoritmos da Aula 08 (Mochila Cheia, Substituição de Objeto e Hill Climbing) são melhores que o algoritmo embaralhado da Aula 07 em termos de consistência e resultados. A Substituição de Objeto e o Hill Climbing melhoram as soluções iniciais de forma iterativa, o que levou a melhores resultados, especialmente na Execução 3, mesmo que levem um pouco mais de tempo. O algoritmo embaralhado da Aula 07 depende mais da sorte e tem resultados mais variáveis. No geral, o Hill Climbing explora melhor as possibilidades, enquanto a Mochila Cheia é mais rápida, mas menos eficiente.
