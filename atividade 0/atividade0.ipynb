### Elabore um gráfico que relaciona a complexidade do problema

```python
import matplotlib.pyplot as plt
```

```python
tamanho_matriz = ['300x300','900x900','1300x1300']
lista_parte1 = [2.50,69.31,232.45]
lista_parte2 = [0.202707,5.90656,18.3434]
lista_parte3 = [0.193095,5.82324,17.964]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(tamanho_matriz, lista_parte1, marker='o', label='Parte 1')
plt.plot(tamanho_matriz, lista_parte2, marker='o', label='Parte 2')
plt.plot(tamanho_matriz, lista_parte3, marker='o', label='Parte 3')

plt.title('Tamanho da Matriz vs tempo de execução')
plt.xlabel('Tamanho da Matriz')
plt.ylabel('Tempo de execução (s)')
plt.legend()
plt.grid(True)

plt.show()
```
