### Elabore um gráfico que relaciona a complexidade do problema

```python
import matplotlib.pyplot as plt
```

```python
tamanho_matriz = ['300x300','900x900','1300x1300']
lista_python = [5.16,160.22,470.45]
lista_cmoremore = [0.145593,4.03188,4.01843]
lista_cmoremore_com_paralelo = [0.112565,1.15729,3.53235]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(tamanho_matriz, lista_python, marker='o', label='Python')
plt.plot(tamanho_matriz, lista_cmoremore, marker='o', label='C++')
plt.plot(tamanho_matriz,lista_cmoremore_com_paralelo, marker='o', label='C++ com paralelismo')

plt.title('Tamanho da Matriz vs tempo de execução no Cluster')
plt.xlabel('Tamanho da Matriz')
plt.ylabel('Tempo de execução (s)')
plt.legend()
plt.grid(True)

plt.show()
```

### Discuta o impacto do ambiente HPC no desempenho das operações e como o paralelismo pode escalar em um cluster de múltiplos nós.


O ambiente de HPC é ideal para tarefas intensivas, pois distribui o trabalho entre vários nós dentro de um cluster e isso acelera a execução. No gráfico é possível ver que o código C++ com paralelismo se beneficia muito do HPC, executando bem mais rápido que Python ou C++ sem paralelismo. O paralelismo permite que diferentes partes de uma matriz sejam processadas ao mesmo tempo, escalando o desempenho conforme o número de nós e threads aumenta.


Quando comparamos a execução em um cluster HPC com um PC normal, a principal diferença é que o HPC consegue escalar melhor as tarefas. Em um PC normal, temos recursos limitados, então tarefas grandes ficam mais lentas. Já no HPC, a tarefa é dividida entre vários nós, cada um com suas CPUs e memória, o que permite que o trabalho seja feito em paralelo e mais rápido. Isso é ótimo para coisas que exigem muito poder de processamento, como multiplicação de matrizes grandes.
