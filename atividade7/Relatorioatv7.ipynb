Para a tarefa 1

```python
import matplotlib.pyplot as plt

pesos = [373, 239, 600]
tempos = [2.34295, 118.411, 10.2988]
entradas = ['Entrada 1', 'Entrada 2', 'Entrada 4']

plt.figure(figsize=(10, 6))
plt.plot(pesos, tempos, marker='o', linestyle='-', color='b')

for i, txt in enumerate(entradas):
    plt.annotate(txt, (pesos[i], tempos[i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.title('Peso Ocupado (Kg) vs Tempo de Execução (segundos)')
plt.xlabel('Peso Ocupado (Kg)')
plt.ylabel('Tempo de Execução (segundos)')
plt.grid(True)

plt.show()
```

### Para a tarefa 2

```python
pesos = [532, 738, 559]
tempos = [0.456149, 2.3984, 0.0630381]
entradas = ['Entrada 1', 'Entrada 2', 'Entrada 4']

plt.figure(figsize=(10, 6))
plt.plot(pesos, tempos, marker='o', linestyle='-', color='b')

for i, txt in enumerate(entradas):
    plt.annotate(txt, (pesos[i], tempos[i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.title('Peso Ocupado (Kg) vs Tempo de Execução (segundos)')
plt.xlabel('Peso Ocupado (Kg)')
plt.ylabel('Tempo de Execução (segundos)')
plt.grid(True)

plt.show()
```

A Entrada 2 é notavelmente mais complexa para ambas as heurísticas, levando a um tempo de execução mais longo e, no caso da Atividade 1, um resultado demorado. Isso reflete que certas distribuições de itens podem ser mais difíceis de resolver de forma eficiente. Se o objetivo é maximizar o valor alcançado, a Atividade 1 pode ser mais confiável, embora mais demorada. Para aplicações onde o tempo é crítico e solução  é aceitável, a Atividade 2 seria a escolha preferida.
