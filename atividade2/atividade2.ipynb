#### 1. Desempenho do código é afetado pela modificação do aumento do número de iterações de convolução

````python
import matplotlib.pyplot as plt

iterations = [15, 30, 50, 100]
tempo_limitado = [0.872479, 1.83245, 5.57936, 8.70638]
tempo_buffado = [0.248958, 0.473883, 0.756403, 1.50748]

plt.figure(figsize=(10, 6))
plt.plot(iterations, tempo_limitado, marker='o', linestyle='-', label='convolucao_limitado.slurm')
plt.plot(iterations, tempo_buffado, marker='o', linestyle='-', label='convolucao_buffado.slurm')
plt.title('Tempo de Execução vs Número de Iterações de Convolução')
plt.xlabel('Número de Iterações')
plt.ylabel('Tempo de Execução (segundos)')
plt.legend()
plt.grid(True)
plt.show()
````

Quando aumentamos o número de iterações no código, a complexidade computacional também aumenta, o que faz com que o tempo de execução seja maior. Com recursos limitados, como no caso do convolucao_limitado.slurm, o tempo de execução cresce bastante à medida que as iterações aumentam. Por outro lado, quando usamos mais recursos, como no convolucao_buffado.slurm, o impacto desse aumento de iterações é bem menor. Isso mostra que é muito importante ajustar os recursos no SLURM de acordo com a complexidade da tarefa, para garantir que o código seja executado de forma mais eficiente.








#### 2.Analise do impacto da alocação de mais recursos, como o número de CPUs e a quantidade de memória, no tempo de execução

````python
iterations = [15, 30, 50, 100]

tempo_intermediario = [0.463, 0.900, 1.637, 3.150] # 1 CPU e 512MB de memória
tempo_limitado = [0.872479, 1.83245, 5.57936, 8.70638] # 2 CPUs e 768MB de memória
tempo_buffado = [0.248958, 0.473883, 0.756403, 1.50748] # 4 CPUs e 1GB de memória

plt.figure(figsize=(10, 6))
plt.plot(iterations, tempo_limitado, marker='o', linestyle='-', label='convolucao_limitado.slurm')
plt.plot(iterations, tempo_intermediario, marker='o', linestyle='-', label='convolucao_intermediario.slurm')
plt.plot(iterations, tempo_buffado, marker='o', linestyle='-', label='convolucao_buffado.slurm')
plt.title('Tempo de Execução vs Número de Iterações de Convolução')
plt.xlabel('Número de Iterações')
plt.ylabel('Tempo de Execução (segundos)')
plt.legend()
plt.grid(True)
plt.show()
````


Os dados mostram que, quando aumentamos os recursos, como o número de CPUs e a quantidade de memória, o tempo de execução diminui bastante, principalmente quando o número de iterações aumenta. Isso faz perceber que, com mais recursos, o código roda mais rápido, mesmo quando a tarefa fica mais complicada. Portanto, alocar mais recursos ajuda a melhorar o desempenho do algoritmo, tornando a execução mais eficiente, mesmo em situações de maior complexidade.

#### 3. Como os comandos sinfo, squeue, sprio, srun, sstat, e strigger podem ser utilizados para monitorar e gerenciar jobs no cluster

**sinfo**: Exibe o estado dos nós e das filas no cluster. Isso é útil para verificar quais recursos estão disponíveis antes de submeter um job.

**squeue**: Lista os jobs atualmente em execução, pendentes ou finalizados no cluster. Permite monitorar o progresso dos seus jobs e ver a fila de espera.

**sprio**: Mostra a prioridade dos jobs. Isso pode ser útil para entender por que um job específico está demorando para ser executado.

**srun**: Executa um job interativamente no terminal, permitindo observar o comportamento do programa em tempo real. Isso é útil para testar diferentes configurações de recursos antes de submeter um job com sbatch

**sstat**: Exibe informações sobre o status de jobs ativos, como uso de memória e tempo de CPU. Isso ajuda a monitorar o consumo de recursos durante a execução de um job.
```
sstat --jobs=<JOB_ID>
```

**strigger**: Define gatilhos automáticos que executam ações quando certos critérios são atendidos. Isso é útil para receber notificações automáticas sobre o status de seus jobs.
